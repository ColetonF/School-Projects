```python
plt.figure(figsize=(10,10))
plt.xlabel('days')

ax1 = myData.Cideal.plot(color='blue', grid=True, label='work completed - ideal')
ax3 = myData.Cactual.plot(color='red', grid=True, secondary_y=True, label='work completed - actual')

ax2 = myData.Lideal.plot(color='green', grid=True, label='work left - ideal')
ax4 = myData.Lactual.plot(color='orange', grid=True, secondary_y=True, label='work left - actual')

h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax1.get_legend_handles_labels()
h3, l3 = ax1.get_legend_handles_labels()
h4, l4 = ax1.get_legend_handles_labels()

#plt legend 
```

```python
fig, (ax1,ax2)= plt.subplots(1,2, figsize=(15,7))


ax1.set_xlabel('days')


ax1.plot(myData.Cideal, color='pink', label='work completed - ideal')
ax3=ax1.twinx()
ax1.plot(myData.Cactual, color='purple', label='work completed - actual')

ax2.plot(myData.Lideal, color='pink', label='work to be done - ideal')
ax4=ax2.twinx()
ax2.plot(myData.Lactual, color='purple', label='work to be done - actual')

ax1.set_title('Burn Up')
ax2.set_title('Burn Down')

h1, l1 = ax1.get_legend_handles_labels()
h3, l3 = ax3.get_legend_handles_labels()
ax3.legend(h1+h3,l1+l3,loc=2)
h2, l2 = ax2.get_legend_handles_labels()
h4, l4 = ax4.get_legend_handles_labels()
ax4.legend(h1+h3,l1+l3,loc=2)


plt.show()
```
